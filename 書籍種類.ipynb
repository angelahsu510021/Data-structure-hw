```python
import sys

class Book:
  def _init_(self):
    self.book=''
    self.category=''
    self.price=0
    self.next=None

ptr=None
current=None
prev=None

head=Book()
head.next=None

#副程式輸入
def insert_f():
  global ptr
  global head
  global current
  global prev

  ptr=Book()
  ptr.next=None
  ptr.book=input('書籍名稱:')
  ptr.category=input('書籍種類:')
  ptr.price=eval(input('書籍價格:'))
  print()

  prev=head
  current=head.next
  while current != None and current.price >= ptr.price:
    prev=current
    current=current.next
  ptr.next=current
  prev.next=ptr

#副程式刪除
def delete_f():
  global head
  global current
  global prev

  del_book=''
  if head.next == None:
    print('沒有書籍紀錄\n')
  else:
    del_book=input('刪除書籍名稱:')
    prev=head
    current=head.next
    while current != None and del_book != current.book:
      prev=current
      current=current.next
    if current != None:
      prev.next=current.next
      current=None
      print('\n 書籍 %s 已被刪除 \n' % del_book)
    else:
      print('\n 找不到此本 %s 書籍\n'% del_book)

#副程式呈現
def display_f():
  global head
  global current

  count=0
  if head.next == None:
    print('沒有書籍紀錄\n')
  else:
    print('%-15s %-15s %-15s' % ('書籍','類別','價錢'))
    for i in range(25):
      print('-',end='')
    print()
    current=head.next
    while current != None:
      print('%-17s %-17s %-15d' % (current.book,current.category,current.price))
      count=count+1
      current=current.next
    for i in range(25):
      print('-',end='')
    print()
    print('總共找到 %d 筆資料\n' % count)


#主程式選擇數字去呼叫副程式
def main():
  option=0
  while True:
    print('*******    書籍資料    *******')
    print('    1 輸入           ')
    print('    2 刪除           ')
    print('    3 呈現           ')
    print('    4 離開           ')
    print('***************************************')

    try:
      option=int(input(' 選擇:   '))
    except ValueError:
      print('數字不正確')
      print('再試一次\n')

    print()
    if option == 1:
      insert_f()
    elif option ==2:
      delete_f()
    elif option ==3:
      display_f()
    elif option == 4:
      sys.exit(0)

main()
#輸入4會錯誤那個理會，那是類別的問題喔喔
```
